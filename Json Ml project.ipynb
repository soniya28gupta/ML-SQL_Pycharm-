```python
# Try reading the SQL file using latin1 encoding
with open("world.sql", "r", encoding='latin1') as f:
    sql_commands = f.read()

# Execute each SQL command
with connection.cursor() as cursor:
    for command in sql_commands.split(";"):
        if command.strip():
            try:
                cursor.execute(command)
            except Exception as e:
                print(f"Error executing:\n{command[:100]}...\n{e}")
                
print("SQL script loaded successfully.")

```
